Generate a dataframe having the following columns name{Jan to dec} with a set index of 1 to 30

generate a random value to fill the dataframe ranging from 5000 to 500000 using seed number 10

find the out the folowing

1.the mean score of [Jan, May, Aug and sep]

2.the maximum sales for each month

3.the 20th,50th,75th percentile for [Feb, June, Dec]

4.the difference between the total sale of june compared to july

5.find out if the sale is stable or not stable

```python
import numpy as np
import pandas as pd
```

```python
colum = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
idexx = np.arange(1,31)
np.random.seed(10)
dataa = np.random.randint(5000,500000,360)
dataa = dataa.reshape(30,12)

saless = pd.DataFrame(index = idexx, data = dataa, columns = colum)
saless
```

```python
pd.DataFrame.mean(saless[['Jan','May','Aug','Sep']])
```

```python
print('the maximum sales for each month'.upper())
pd.DataFrame.max(saless[['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']])
```

```python
np.percentile(saless[['Feb','Jun','Dec']],20)
```

```python
np.percentile(saless[['Feb','Jun','Dec']],50)
```

```python
np.percentile(saless[['Feb','Jun','Dec']],75)
```

```python
jun = np.sum(saless['Jun']) #- np.sum(saless[['Jul']])
jul = np.sum(saless['Jul'])
print(f'total sale (Jun): {jun}'.title())
print(f'total sale (Jul): {jul}'.title())
print(f'the difference between the total sale of june compared to july is: {jun-jul}'.title())


```
